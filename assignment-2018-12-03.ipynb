```python
import tensorflow as tf
```

```python
def train_input_fn(features, labels, batch_size):
    # Convert the inputs to a Dataset.
    dataset = tf.data.Dataset.from_tensor_slices((dict(features), labels))

    # Shuffle, repeat, and batch the examples.
    dataset = dataset.shuffle(1000).repeat().batch(batch_size)

    # Return the dataset.
    return dataset
```

```python
def eval_input_fn(features, labels, batch_size):
    """An input function for evaluation or prediction"""
    features=dict(features)
    if labels is None:
        # No labels, use only features.
        inputs = features
    else:
        inputs = (features, labels)

    # Convert the inputs to a Dataset.
    dataset = tf.data.Dataset.from_tensor_slices(inputs)

    # Batch the examples
    assert batch_size is not None, "batch_size must not be None"
    dataset = dataset.batch(batch_size)

    # Return the dataset.
    return dataset
```

```python
x_col = list(df_new.columns)
x_col.remove("sal_>50K")
x_col.remove("sal_<=50K")
y_col = ["sal_>50K"]

val = math.floor(len(df_new)/10) *  8

X_train = df_new[x_col][:val]
y_train = df_new[y_col][:val]
X_test = df_new[x_col][val:]
y_test = df_new[y_col][val:]
y_test = y_test.astype(int)
y_train = y_train.astype(int)
# Feature columns describe how to use the input.
# We are adding one numeric feature for each column of the training data.

import statsmodels.stats.api as sms
def NeuralNetwork(X_train,y_train,X_test,y_test):
    my_feature_columns = []
    for key in X_train.keys():
        fc = tf.feature_column.numeric_column(key=key)
        print(key, fc)
        my_feature_columns.append(fc)

    # Build 2 hidden layer DNN with 10, 10 units respectively.
    classifier = tf.estimator.DNNClassifier(
        feature_columns = my_feature_columns,

        # Two hidden layers of 10 nodes each.
        hidden_units = [10, 10],

        # The model must choose between 3 classes.
        n_classes=3,

        ## We can also set the directory where model information will be saved.
        ##model_dir='models/iris'
        )
    
    classifier.train(input_fn = lambda:train_input_fn(X_train, y_train, 100),steps = 1000)
    eval_result = classifier.evaluate(
        input_fn = lambda:eval_input_fn(X_test, y_test, 100))
    return eval_result
acc=[]
for i in range(20):
    eval_result=NeuralNetwork(X_train,y_train,X_test,y_test)
    acc.append(eval_result['accuracy'])
#print('\nTest set accuracy: {accuracy:0.3f}\n'.format(eval_result))
```
